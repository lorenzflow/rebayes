# Rotating moons 

```python
import numpy as np
import matplotlib.pyplot as plt
from rebayes.utils import datasets
```

```python
%config InlineBackend.figure_format = "retina"
%load_ext autoreload
%autoreload 2
```

```python
n_train, n_test = 200, 100
n_rotations = 4
noise = 0.3
```

```python
train, test = datasets.make_rotating_moons(n_train, n_test, n_rotations, max_angle=180, noise=0.2)
X_train, y_train, rotations_train = train
```

```python
fig, axs = plt.subplots(2, 2)
axs = axs.ravel()

for i, ax in enumerate(axs):
    slice_ix = slice(i * n_train, n_train * (i + 1))
    rotation = rotations_train[slice_ix.start]
    deg = np.rad2deg(rotation)
    ax.scatter(*X_train[slice_ix].T, c=y_train[slice_ix], cmap="Dark2")
    ax.axis("equal")
    ax.axis("off")
    ax.set_title(f"angle: {deg:0.2f}")
plt.tight_layout()
```
