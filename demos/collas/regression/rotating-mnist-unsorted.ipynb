## Plot results

```python
dir_output = "../../figures/"
task_name = "gradually-rotating"
```

```python
fig, ax = plt.subplots(figsize=(7, 4))

err_smooth_rsgd_adam = output_rsgd_adam["window-metric"]
err_smooth_lofi = output_lofi["window-metric"]

plt.plot(err_smooth_lofi[10:], label="lofi")
plt.plot(err_smooth_rsgd_adam[10:], label="rsgd-adam")

plt.grid(alpha=0.3)

vmin = np.quantile(err_smooth_lofi[20:], 0.00)
vmax = np.quantile(err_smooth_rsgd_adam[20:], 0.99)
plt.legend()
plt.title("window-10 RMSE")
plt.xlabel("num. observation")

namefile = "window-rmse.pdf"
namefile = os.path.join(dir_output, f"{task_name}-{namefile}")
plt.savefig(namefile)
# plt.ylim(vmin, vmax);
```

```python
fig, ax = plt.subplots(figsize=(7, 4))

err_smooth_lofi = output_lofi["osa-metric"]
err_smooth_rsgd_adam = output_rsgd_adam["osa-metric"]

err_smooth_lofi = np.convolve(err_smooth_lofi, kernel, mode=kmode)
err_smooth_rsgd_adam = np.convolve(err_smooth_rsgd_adam, kernel, mode=kmode)

plt.plot(err_smooth_lofi, label="lofi", zorder=1)
plt.plot(err_smooth_rsgd_adam, label="rsgd-adam", zorder=0)

plt.grid(alpha=0.3)

vmin = np.quantile(err_smooth_rsgd_adam, 0.00)
vmax = np.quantile(err_smooth_rsgd_adam, 0.99)
plt.legend()
plt.title(f"step-ahead RMSE (smooth k{ksize})")
plt.xlabel("num. observation")

namefile = "ose-rmse.pdf"
namefile = os.path.join(dir_output, f"{task_name}-{namefile}")
plt.savefig(namefile)
# plt.ylim(vmin, vmax);
```

```python
fig, ax = plt.subplots(figsize=(7, 4))

metric_lofi = output_lofi["nlpd"]
metric_rsgd_adam = output_rsgd_adam["nlpd"]

metric_lofi = np.convolve(metric_lofi, kernel, mode=kmode)
metric_rsgd = np.convolve(metric_rsgd_adam, kernel, mode=kmode)

plt.plot(metric_lofi, label="lofi", zorder=1)
plt.plot(metric_rsgd_adam, label="rsgd-adam", zorder=0)

plt.title(f"step-ahead NLPD (smooth k{ksize})")
plt.xlabel("num. observation")

plt.legend()
plt.grid(alpha=0.3)

namefile = "ose-nlpd.pdf"
namefile = os.path.join(dir_output, f"{task_name}-{namefile}")
plt.savefig(namefile)
```

```python
fig, ax = plt.subplots(figsize=(7, 4))

err_smooth_lofi = output_lofi["nldp_window"]
err_smooth_rsgd_adam = output_rsgd_adam["nldp_window"]

plt.plot(err_smooth_lofi[10:], label="lofi")
plt.plot(err_smooth_rsgd_adam[10:], label="rsgd-adam")

plt.grid(alpha=0.3)

plt.legend()
plt.title("window-10 NLPD")
plt.xlabel("num. observation")

vmin = np.quantile(err_smooth_lofi[100:], 0.00)
vmax = np.quantile(err_smooth_rsgd_adam[100:], 0.999)
plt.ylim(vmin, vmax)

namefile = "window-nlpd.pdf"
namefile = os.path.join(dir_output, f"{task_name}-{namefile}")
plt.savefig(namefile)
```
